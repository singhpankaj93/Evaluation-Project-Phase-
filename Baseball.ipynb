```python
#List the numerical variable
Numerical_features=[feature for feature in df.columns if((df[feature].dtypes!='O') & (feature not in ["W"]))]
len(Numerical_features)
```

```python
#print the Numerical feature
df[Numerical_features].head()
```

```python
#Count the Discrete numerical variable
discrete_feature=[feature for feature in Numerical_features if len(df[feature].unique())<25]
print("Discrete variable count: {}".format(len(discrete_feature)))
```

```python
df[discrete_feature].head()
```

```python
# count the continous features
Continous_features=[feature for feature in Numerical_features if feature not in discrete_feature+['W']]
print("continous feature count: {}".format(len(Continous_features)))
```

```python
df[Continous_features].head()
```

```python
sns.countplot(x="W",data=df)
plt.show()
```

```python
plt.figure(figsize=(20,60),facecolor="white")
plot_number=1
for continous_feature in discrete_feature:
    ax=plt.subplot(12,3,plot_number)
    sns.displot(df[discrete_feature])
    plt.xlabel(discrete_feature)
    plot_number+=1
plt.show()
```

```python
#distribution of continous numerical feature
#plot univariant distribution of continous observation
plt.figure(figsize=(20,60),facecolor="white")
plot_number=1
for continous_feature in Continous_features:
    ax=plt.subplot(12,3,plot_number)
    sns.displot(df[continous_feature])
    plt.xlabel(continous_feature)
    plot_number+=1
plt.show()
```
